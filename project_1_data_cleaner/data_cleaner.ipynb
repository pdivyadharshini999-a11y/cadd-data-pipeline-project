```python
import pandas as pd
```

```python
data = {
    "Molecule_Name": ["Aspirin", "Paracetamol", "Ibuprofen", "Unknown_X", None, "Caffeine", "Nicotine"],
    "SMILES": ["CC(=O)OC1=CC=CC=C1C(=O)O", "CC(=O)NC1=CC=CC=C1O", "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O", None, "CN1C=NC2=C1C(=O)N(C(=O)N2C)C", "C1=CC(=CN=C1)C2CCCN2C", "INVALID_DATA"],
    "Solubility_Score": [1.2, None, 0.8, -2.5, 1.5, None, 9999.0],
    "Binding_Affinity": [75, 60, None, 82, 55, 90, -1]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df[df["Solubility_Score"]<100]
```

```python
df["Molecule_Name"]=df["Molecule_Name"].fillna(df["Molecule_Name"].mode()[0])
df["SMILES"] = df["SMILES"].fillna("Unknown")
df["Solubility_Score"]=df["Solubility_Score"].fillna(df["Solubility_Score"].mean())
df["Binding_Affinity"]=df["Binding_Affinity"].fillna(df["Binding_Affinity"].median())
```

```python
df.to_csv("clean_data.csv",index=False)
```
